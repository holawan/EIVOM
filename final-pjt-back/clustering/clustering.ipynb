## hierarchical

```python
from sklearn.model_selection import KFold, StratifiedKFold, train_test_split
from sklearn.preprocessing import Normalizer, StandardScaler, RobustScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
import pandas as pd
```

```python
tmp = pd.read_csv('./preprocessing_movie.csv')
movies = tmp[['id', 'genres']]
```

```python
movies = tmp[['id', 'genres']]
movies['genres'] = movies['genres'].str.strip('[]')
genre_list = [12,14,16,18,27,28,35,36,37,53,80,99,878,9648,10402,10749,10751,10752,10770]
import re
cols = ['id','genres']

def filterGenre(word, strings) :
    if re.search(word,strings) :
        val_return = 1 
    else :
        val_return = 0 
    return val_return 
for i in genre_list :
    val = movies.apply(lambda x: filterGenre(str(i),x['genres']),axis=1)
    movies[i] = val 
```

```python
dummy2 = movies.iloc[:, 2:] # 첫 2개열만
```

```python
dummy2
```

```python
### Dendrogram : Euclidean

aug_affinity = 'euclidean'
aug_methods = ['ward', 'complete', 'average', 'single']

for aug_method in aug_methods:
    plt.figure(figsize=(8, 3))
    plt.title('affinity : {} / method : {}'.format(aug_affinity, aug_method))
    hier_link = sch.linkage(dummy2, metric=aug_affinity, method=aug_method)
    hier_dend = sch.dendrogram(hier_link)
    plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
### Dendrogram : Cosine

aug_affinity = 'cosine'
aug_methods = ['complete', 'average', 'single']

for aug_method in aug_methods:
    plt.figure(figsize=(8, 3))
    plt.title('affinity : {} / method : {}'.format(aug_affinity, aug_method))
    hier_link = sch.linkage(dummy2, metric=aug_affinity, method=aug_method)
    hier_dend = sch.dendrogram(hier_link)
    plt.show()
```

```python
### Clustering

n_cluster = 6
array_matrix = dummy2.values
aug1 = 'euclidean'
aug2 = 'ward'

clustering = AgglomerativeClustering(n_clusters=n_cluster,
                                     affinity=aug1, linkage=aug2)

array_cluster = clustering.fit_predict(array_matrix)
dummy2['cluster'] = array_cluster
dummy2.head(10)
```

```python
### Clustering : Pie Plot
df_pie = dummy2.groupby('cluster').agg({14:'count'}).reset_index()
df_pie.rename(columns={14:'count'}, inplace=True)

explodes = [0, .005, .005]

plt.figure(figsize=(5,5))
plt.pie(df_pie['count'], labels=df_pie['cluster'], autopct='%.2f%%', 
        explode=explodes)
plt.legend()
plt.show()

df_pie
```

```python
### Aggregate By Clusters

cols = ['cluster']
dummy2.columns
cols_val = dummy2.columns.tolist()[:-1]
df_agg = dummy2.groupby(cols)[cols_val].mean()
df_agg
```

```python
### Heatmap : Aggregate By Clusters

plt.figure(figsize=(6, 6))
sns.heatmap(df_agg.T, cmap='coolwarm', vmax=4, vmin=0, linewidths=.01,
            annot=True, fmt='.2f')
plt.title('Average Rating By Clusters & Genres', fontsize=12)
plt.show()
```
